```python
import pandas as pd
import yaml
 
try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("Yaml configuration file not found!")

boardgames_df = pd.read_csv(config['input_data']['file'])
```

```python
boardgames_df
```

```python
boardgames_df = boardgames_df.drop(columns=["release_year", "rank_overall", "owned", "fans", "page_views", "std_deviation",	"comments", "monthly_plays", "previously_owned", "for_trade", "want_trade", "rating_1", "rating_2", "rating_3", "rating_4", "rating_5", "rating_6", "rating_7", "rating_8", "rating_9", "rating_10", "mechanics", "families", "designers", "artists", "publishers", "solo_designers", "developers", "graphic_designers", "sculptors", "editors", "writers", "insert_designers", "rank_strategy", "rank_thematic", "rank_family", "rank_war", "rank_customizable", "rank_abstract", "rank_party", "rank_childrens"])
boardgames_df = boardgames_df.rename(columns={"row_id": "game_id"})
boardgames_df["categories"] = boardgames_df["categories"].fillna("Unknown")
#boardgames_df = boardgames_df["categories"].replace("NaN", "Unknown")
boardgames_df
```

```python
boardgames_df['categories'].isna().sum()
```

```python
categories_df = boardgames_df['categories'].str.split(";",expand=True)

display(categories_df.head())
#categories_df.info()
categories_list = []

for col in range(categories_df.shape[1]):
    for row in range(categories_df.shape[0]):
        if categories_df.iloc[row, col] != 'None' and categories_df.iloc[row, col] not in categories_list:
            categories_list.append(categories_df.iloc[row, col])

categories_list
```

```python
categories_names_df = pd.DataFrame({"category_id": [ i + 1 for i in range(len(categories_list))], "name": categories_list})
display(categories_names_df)
categories_names_df.to_csv("../data/clean/categories.csv", sep=";", encoding="utf-8", index=False)
```

```python
playtime_df = boardgames_df[['min_playtime', 'max_playtime']].drop_duplicates()
playtime_df['playtime_id'] = [i + 1 for i in range(playtime_df.shape[0])]
playtime_df = playtime_df[['playtime_id', 'min_playtime', 'max_playtime']]
display(playtime_df)
playtime_df.to_csv("../data/clean/playtime.csv", sep=";", encoding="utf-8", index=False)
```

```python
boardgames_df["minimum_age"] = boardgames_df["minimum_age"].replace(0, 4)
players_df = boardgames_df[['min_players', 'max_players','minimum_age']].drop_duplicates()
players_df['player_id'] = [i + 1 for i in range(players_df.shape[0])]
players_df = players_df[['player_id','min_players', 'max_players','minimum_age']]
display(players_df)
players_df.to_csv("../data/clean/players.csv", sep=";", encoding="utf-8", index=False)
```

```python
merged_df = pd.merge(playtime_df, boardgames_df, left_on=['min_playtime','max_playtime'], right_on=['min_playtime','max_playtime'], how='right')

merged_2_df = pd.merge(players_df, merged_df, left_on=['min_players', 'max_players','minimum_age'], right_on=['min_players', 'max_players','minimum_age'], how='right')

games_df = merged_2_df.drop(['min_players', 'max_players', 'minimum_age', 'min_playtime', 'max_playtime'], axis=1)

games_df = games_df[['game_id', 'boardgame', 'avg_rating', 'num_ratings', 'complexity', 'wishlisted', 'total_plays', 'amazon_price', 'url', 'description', 'player_id', 'playtime_id']]
display(games_df.head())
games_df.to_csv("../data/clean/games.csv", sep=";", encoding="utf-8", index=False)
```

```python
def recommendation_boardgames(boardgames_df):
    # Get user preferences and convert to integers
    playtime = int(input('Please enter the number of minutes you wish to play: '))
    number_players = int(input('Please enter the number of players: '))
    min_age = int(input('Please enter the age of the youngest player: '))
    complexity = float(int(input('Please enter the difficulty level you wish for the game from 1 to 4 (1 - Easy; 2 - Medium; 3 - Hard; 4 - Very Hard): ')))

    # Filter the dataframe using conditions
    filtered_df = boardgames_df[
        (boardgames_df['min_playtime'] <= playtime) &
        (boardgames_df['max_playtime'] <= playtime) &
        (boardgames_df['min_players'] <= number_players) &
        (boardgames_df['max_players'] >= number_players) &
        (boardgames_df['minimum_age'] <= min_age) &
        (boardgames_df['complexity'] <= complexity+1) &
        (boardgames_df['complexity'] >= complexity)
    ]

    filtered_df_ranked = filtered_df.sort_values(by="avg_rating", ascending=False, inplace=False)

    result = filtered_df_ranked.head(5)
    
    return result
```

```python
recommendation_boardgames(boardgames_df)
```
