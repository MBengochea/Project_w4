```python
boardgames_df = boardgames_df.drop(columns=["release_year", "rank_overall", "owned", "fans", "page_views", "std_deviation",	"comments", "monthly_plays", "previously_owned", "for_trade", "want_trade", "rating_1", "rating_2", "rating_3", "rating_4", "rating_5", "rating_6", "rating_7", "rating_8", "rating_9", "rating_10", "mechanics", "families", "designers", "artists", "publishers", "solo_designers", "developers", "graphic_designers", "sculptors", "editors", "writers", "insert_designers", "rank_strategy", "rank_thematic", "rank_family", "rank_war", "rank_customizable", "rank_abstract", "rank_party", "rank_childrens"])
boardgames_df
```

```python
boardgames_df.to_csv("../data/clean/boardgames_df_cleaned.csv", index=False)
```

```python
# --- User Inputs ---
playtime = int(input(" Desired playtime (minutes): "))
number_players = int(input(" Number of players: "))
min_age = int(input(" Youngest player age: "))
complexity = float(input(" Desired complexity (1–5): "))

# --- Scoring Function Based on Real Correlations ---
def score_game(row):
    score = 0

    # Complexity (weight 2.5)
    diff = abs(complexity - row['complexity'])
    if diff <= 0.5:
        score += 2.5
    elif diff <= 1.0:
        score += 1.5
    elif diff <= 1.5:
        score += 0.5

    # Max Playtime (weight 2.0)
    if row['min_playtime'] <= playtime <= row['max_playtime']:
        score += 2.0
    elif abs(playtime - row['max_playtime']) <= 15:
        score += 1.0

    # Min Playtime (weight 1.5)
    if abs(playtime - row['min_playtime']) <= 15:
        score += 1.5

    # Min Age (weight 1.5)
    if min_age >= row['minimum_age']:
        score += 1.5
    elif min_age >= row['minimum_age'] - 2:
        score += 0.5

    # Min Players (negative weight)
    if number_players < row['min_players']:
        score -= 1.0

    # Max Players (negative weight)
    if number_players > row['max_players']:
        score -= 0.5

    return score

# --- Apply Scoring ---
df = boardgames_df.copy()
df['score'] = df.apply(score_game, axis=1)
df = df[df['score'] > 0]
df = df.sort_values(by=['score', 'avg_rating'], ascending=False)

# --- Show Top Recommendations ---
print("\n Top 5 Correlation-Calibrated Recommendations:")
print(df[['boardgame', 'score', 'avg_rating', 'min_playtime', 'max_playtime', 'minimum_age', 'complexity']].head(5))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Only include numeric columns relevant to scoring
corr_matrix = boardgames_df[[
    'min_playtime', 'max_playtime',
    'min_players', 'max_players',
    'minimum_age', 'complexity',
    'avg_rating', 'num_ratings', 'total_plays'
]].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("📊 Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

complexity has a moderate correlation with avg_rating (0.54), not strong

min_age and max_playtime are mildly predictive (0.35 and 0.36)

min_players and max_players are negatively correlated, so they should be penalized if mismatched

Complexity is prioritized because it strongly predicts rating

Player count matters more than playtime or age
