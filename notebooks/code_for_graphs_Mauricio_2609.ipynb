```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

# --- Load config and dataset ---
try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("Yaml configuration file not found!")

boardgames = pd.read_csv(config['input_data']['file'])


# --- Correlation matrix for key metrics ---
cols = ["complexity", "amazon_price", "avg_rating", "total_plays"]
corr = boardgames[cols].corr()

# --- Pure orange heatmap ---
plt.figure(figsize=(6, 5))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="Oranges",  # built-in sequential orange palette
    vmin=0, vmax=1,  # restrict to positive correlations
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation Between Game Metrics", fontsize=14)
plt.tight_layout()
plt.show()
```
